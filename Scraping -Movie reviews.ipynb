```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
def current_movies_in():
    l=[]
    response = requests.get('https://www.imdb.com/showtimes/?ref_=shlc_sh')
    a = BeautifulSoup(response.text, "html.parser")
    k=a.find_all("a", itemprop="url")
    base="https://www.imdb.com/showtimes/title"
    base1="?ref_=sh_ov_tt"
    for i in range(len(k)):
        x=re.findall('/tt\w+',str(k[i]))
        if x:
            if x[0] not in l:
                l.append(x[0])
    d={}
    for i in l:
        p=[]
        q=[]
        new_link=base+i+base1
        response = requests.get(new_link)
        a = BeautifulSoup(response.text, "html.parser")
        k=a.find_all("a", itemprop="url")
        for j in k:
            if '/ci' in str(j):
                x=re.findall('/ci\w+',str(j))
                if x:
                    p.append(j.text)
        d[i]=p
    base='https://www.imdb.com/title'
    base1='/reviews/?ref_=tt_ql_2'
    link,con,title,date,name,rate,d1=[],[],[],[],[],[],{}
    for i in l:
        new_link=base+i+base1
        response = requests.get(new_link)
        a = BeautifulSoup(response.text, "html.parser")
        rev=a.find_all(class_='text show-more__control')
        rev1=a.find_all(class_='title')
        rev2=a.find_all(class_='review-date')
        rev3=a.find_all(class_='display-name-link')
        rev4=a.find_all('a',itemprop='url')
        rev5=a.find_all('span',class_='rating-other-user-rating')
        for w in rev4:
            d1[i]=w.text
        for p,j,k,n,o in zip(rev,rev1,rev2,rev3,rev5):
            con.append(p.text)
            title.append(j.text)
            date.append(k.text)
            name.append(n.text)
            rate.append(o.text)
        link+=[i]*len(rev5)
    df=pd.DataFrame({'ID':link,'Name':name,'Date':date,'Title':title,'Review':con,'Rating':rate})

    mv=[]
    th=[]
    for i in range(len(df)):
        for j in d1:
            if df['ID'][i]==j:
                mv.append(d1[j])
        for j in d:
            if df['ID'][i]==j:
                th.append(d[j])
    df['Movie Name']=mv
    df['Theatres']=th
    return df
    
```

```python
df=current_movies_in()
df
```

```python
df.to_csv('Movie_Review.csv')
```
