```python
df['Review']=df['Title']+df['Review']
df['Review']
```

```python
df.drop(['Unnamed: 0.1','Unnamed: 0','Title'],axis=1,inplace=True)
```

```python
#removing white space from ratings
import re
for i in range(len(df)):
    x=re.sub('\s+','',df['Rating'][i])
    df['Rating'][i]=x
df
```

```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

def preprocess_text(text):
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = text.lower()
    words = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]
    processed_text = ' '.join(words)

    return processed_text

df['Review'] = df['Review'].apply(preprocess_text)
df['Movie Name'] = df['Movie Name'].apply(preprocess_text)
```

# SENTIMENT ANALYSIS

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.sentiment.util import *
from textblob import TextBlob
```

```python
df['Review']=df['Review'].astype('str')
def get_polarity(text):
    return TextBlob(text).sentiment.polarity
df['polarity']=df['Review'].apply(get_polarity)
```

```python
df['sentiment_type']=''
df.loc[df.polarity>0.1,'sentiment_type']='POSITIVE'
df.loc[(df.polarity < 0.1) & (df.polarity >= 0),'sentiment_type']='NEUTRAL'
df.loc[df.polarity<0,'sentiment_type']='NEGATIVE'
```

# GOOD MOVIE RECOMMENDATIONS

```python
def top_nmovies(n):
    pos=df.loc[df['sentiment_type']=='POSITIVE']
    x=pos.groupby('Movie Name').count()
    x.reset_index(inplace=True)
    y=x.sort_values(by='sentiment_type',ascending=False)['Movie Name'][:n]
    return y
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def graph1():
    pos=df.loc[df['sentiment_type']=='POSITIVE']
    x=pos.groupby('Movie Name').count()
    x.reset_index(inplace=True)
    r=x.sort_values(by='Review',ascending=False)
    plt.rc('font', size=20)
    plt.figure(figsize=[25,25])
    z=sns.barplot(x='Review',y='Movie Name',data=r,orient='h')
    z.set(title='MOVIE WITH GOOD REVIEWS')
    for i in z.containers:
        z.bar_label(i,)
```
